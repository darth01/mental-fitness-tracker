```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df1 = pd.read_csv("mental-and-substance-use-as-share-of-disease.csv")
df2 = pd.read_csv("prevalence-by-mental-and-substance-use-disorder.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
data = pd.merge(df1,df2)
data.head()
```

```python
# Missing values in the dataset

data.isnull().sum()
```

```python
# Drop the column

data.dropna(inplace=True)
```

```python
data.head()
```

```python
data.size,data.shape
```

```python
data.drop(['Code'],axis=1,inplace=True)
```

```python
#Column Set


data.set_axis(['Country','Year','Mental disorders','Schizophrenia', 'Bipolar_disorder', 'Eating_disorder','Anxiety','drug_usage','depression','alcohol'], axis='columns', inplace=True)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap - Diseases and Mental Fitness')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.plot()
```

```python
# Pairwise relationships in a dataset

sns.pairplot(data,corner=True)
plt.show()
```

```python
sns.scatterplot(x='Mental disorders',y='alcohol',data=data)
```

```python
fig=px.pie(data,values='Mental disorders',names='Year')
fig.show()
```

```python
# Yearwise variations in mental_disorder of different countries

fig = px.line(data, x="Year", y="Mental disorders", color='Country',markers=True,color_discrete_sequence=['red','blue'],template='plotly_dark')
fig.show()
```

```python
# Transform non-numeric labels to numeric labels

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in data.columns:
  if data[i].dtype == 'object':
    data[i]=encoder.fit_transform(data[i])
```

```python
from sklearn.model_selection import train_test_split

x=data.drop(['Mental disorders'],axis=1)
y=data['Mental disorders']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=20, random_state=10)
```

```python
# Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

lr = LinearRegression().fit(x_train,y_train)

predict1 = lr.predict(x_test)
mse_lr = mean_squared_error(y_test, predict1)
rmse_lr = (np.sqrt(mean_squared_error(y_test, predict1)))
r2_lr = r2_score(y_test, predict1)

print("Linear Regression model performance")
print("Mean square error:",mse_lr)
print("Root mean square error:",rmse_lr)
print("R2 score:",r2_lr)
```

```python
# Random Forest Regression

from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor().fit(x_train,y_train)

predict2 = rfr.predict(x_test)
mse_rfr = mean_squared_error(y_test,predict2)
rmse_rfr = (np.sqrt(mean_squared_error(y_test,predict2)))
r2_rfr = r2_score(y_test,predict2)

print("Random Forest Regression model performance")
print("Mean square error:",mse_rfr)
print("Root mean square error:",rmse_rfr)
print("R2 score:",r2_rfr)
```

```python
# Decision Tree Regressor

from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor().fit(x_train,y_train)

predict3 = dtr.predict(x_test)
mse_dtr = mean_squared_error(y_test, predict3)
rmse_dtr = (np.sqrt(mean_squared_error(y_test, predict3)))
r2_dtr = r2_score(y_test, predict3)

print("Decision Tree Regressor model performance")
print("Mean square error:",mse_dtr)
print("Root mean square error:",rmse_dtr)
print("R2 score:",r2_dtr)
```

```python
# Support Vector Regression

from sklearn.svm import SVR

svr = SVR().fit(x_train,y_train)

predict4 = svr.predict(x_test)
mse_svr = mean_squared_error(y_test, predict4)
rmse_svr = (np.sqrt(mean_squared_error(y_test, predict4)))
r2_svr = r2_score(y_test, predict4)

print("Support Vector Regression model performance")
print("Mean square error:",mse_svr)
print("Root mean square error:",rmse_svr)
print("R2 score:",r2_svr)
```
